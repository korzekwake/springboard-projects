```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
sns.set_context("paper")
%pylab inline
```

```python
cc = pd.read_csv("risk_factors_cervical_cancer.csv", na_values = '?')
cc_sm = cc[['Age', 'Number of sexual partners', 'First sexual intercourse', 'Num of pregnancies', 'Smokes',
               'Hormonal Contraceptives', 'IUD', 'STDs', 'STDs:condylomatosis', 'STDs:cervical condylomatosis', 
            'STDs:vaginal condylomatosis', 'STDs:vulvo-perineal condylomatosis', 'STDs:syphilis', 
            'STDs:pelvic inflammatory disease', 'STDs:genital herpes', 'STDs:molluscum contagiosum', 'STDs:AIDS', 
            'STDs:HIV', 'STDs:Hepatitis B', 'STDs:HPV', 'Dx:Cancer']]
cc_sm.columns = ['age', 'num_sex', 'fir_sex', 'num_preg', 'smokes', 'horm_cont', 'IUD', 'STDtot', 'STDc', 'STDcc',
                 'STDvc', 'STDvpc', 'STDs', 'STDpid', 'STDgh', 'STDmc', 'STDaids', 'STDhiv', 'STDhb', 'STDhpv', 'DxC']
cc_sm.to_pickle("cc_sm.pkl")
cc_sm.head()
```

```python
cc_nm = cc_sm
categ = ['smokes','horm_cont', 'IUD', 'STDtot', 'STDc', 'STDcc', 'STDvc', 'STDvpc', 'STDs', 'STDpid', 
         'STDgh', 'STDmc', 'STDaids', 'STDhiv', 'STDhb', 'STDhpv', 'DxC']
for c in categ:
    cc_nm[c]=cc_nm[c].replace(np.nan, 2).astype(int)
cc_nm.describe()
cc_nm.to_pickle("cc_nm.pkl")
```

```python
# 0: neg
# 1: pos
# 2: no response
```

```python
cc_cat = cc_nm.iloc[:,4:]
cc_cont = cc_nm.iloc[:, 0:4]
cc_cont.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
enc=OneHotEncoder(sparse=False)
cc_enc = cc_cont
categ = ['smokes','horm_cont', 'IUD', 'STDtot', 'STDc', 'STDcc', 'STDvc', 'STDvpc', 'STDs', 'STDpid', 
         'STDgh', 'STDmc', 'STDaids', 'STDhiv', 'STDhb', 'STDhpv', 'DxC']
for col in categ:
       data = cc_cat[[col]]
       enc.fit(data)
       # Fitting One Hot Encoding on train data
       temp = enc.transform(cc_cat[[col]])
       # Changing the encoded features into a data frame with new column names
       temp=pd.DataFrame(temp,columns=[(col+"_"+str(i)) for i in range(data[col].nunique())])
       # In side by side concatenation index values should be same
       # Setting the index values similar to the X_train data frame
       temp = temp.set_index(cc_nm.index.values)
       # adding the new One Hot Encoded varibales to the train data frame
       cc_enc = pd.concat([cc_enc,temp],axis=1)
cc_enc.head()
```

```python
cc_enc.rename(columns={'STDcc_1': 'STDcc_2', 'STDaids_1' : 'STDaids_2'}, inplace=True)
```

```python
from sklearn.preprocessing import Imputer
impute_mean = Imputer()
impute_median = Imputer(strategy = 'median')
impute_mode = Imputer(strategy = 'most_frequent')
```

```python
cc_median = pd.DataFrame(impute_median.fit_transform(cc_cont))
cc_median.columns = cc_cont.columns
cc_median = pd.concat([cc_median, cc_enc.iloc[:,4:]], axis = 1)
cc_median.to_pickle("cc_median.pkl")
cc_median.head()
```

```python
cc_median = pd.DataFrame(impute_median.fit_transform(cc_cont))
cc_median.columns = cc_cont.columns
cc_median = pd.concat([cc_median, cc_enc.iloc[:,4:]], axis = 1)
cc_median.to_pickle("cc_median.pkl")
cc_median.head()
```
